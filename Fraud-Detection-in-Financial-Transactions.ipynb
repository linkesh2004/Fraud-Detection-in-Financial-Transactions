```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
transactions_df=pd.read_csv('Fin Fraud dataset.csv')
transactions_df.tail(10)
```

```python
transactions_df.columns
```

```python
# Convert 'Transaction Date' column to datetime format
transactions_df['Transaction Date'] = pd.to_datetime(transactions_df['Transaction Date'])

# Extract year, month, day, hour, and minute features
transactions_df['Year'] = transactions_df['Transaction Date'].dt.year
transactions_df['Month'] = transactions_df['Transaction Date'].dt.month
transactions_df['Day'] = transactions_df['Transaction Date'].dt.day
transactions_df['Hour'] = transactions_df['Transaction Date'].dt.hour
transactions_df['Minute'] = transactions_df['Transaction Date'].dt.minute

# Drop the original 'Transaction Date' column
transactions_df.drop('Transaction Date', axis=1, inplace=True)


transactions_df.head()
```

```python
transactions_df.info()
```

```python
transactions_df.isnull().sum()
```

```python
me=transactions_df['Transaction Amount'].mean()
transactions_df['Transaction Amount'].fillna(me,inplace=True)

me=transactions_df['Customer Age'].mean()
int_age=int(me//1)
transactions_df['Customer Age'].fillna(int_age,inplace=True)

transactions_df.isnull().sum()
```

```python
transactions_df = transactions_df.dropna(how='any',axis=0)

transactions_df.isnull().sum()
```

```python
transactions_df.info()
```

```python
# Split features and target variable
x = transactions_df.drop(['Transaction ID', 'Fraudulent','Merchant Name','Customer Gender','Customer ID' ,
                          'Customer Location','Device Info','IP Address','Transaction Currency','Transaction Frequency'], axis=1)
y = transactions_df['Fraudulent']

x
```

```python
one_hot_encoded_data = pd.get_dummies(x, columns = ['Transaction Type'])
merged=pd.merge(one_hot_encoded_data, y, left_index=True, right_index=True)
merged.to_csv('Preprocessed Fin fraud dataset.csv')
one_hot_encoded_data
```

```python
# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(one_hot_encoded_data, y, test_size=0.25)
```

```python
# Initialize and train the RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
model.fit(x_train, y_train)
```

```python
# Predict on the test set
y_pred = model.predict(x_test)
```

```python
# Evaluate the model
#Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)
```

```python
#Classification Report
classification_rep = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(classification_rep)
```
